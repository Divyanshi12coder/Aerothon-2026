# =====================================
# Feature Engineering
# ========================================

```python
df["CompressorPressureRatio"] = df["P2_Pa"] / df["Pamb_Pa"]
```

```python
df[["P2_Pa","Pamb_Pa","CompressorPressureRatio"]].head()
```

```python
# Compressor Pressure Ratio vs Compressor Health
```

```python
plt.figure(figsize=(8,6))

plt.scatter(df["CompressorPressureRatio"],
            df["CompressorHealth"])

plt.xlabel("Compressor Pressure Ratio")

plt.ylabel("Compressor Health")

plt.title("Compressor Pressure Ratio vs Compressor Health")

plt.grid(True)

plt.show()
```

```python
# Compressor Temperature Rise
```

```python
df["CompressorTempRise"] = df["T2_K"] - df["Tamb_K"]
```

```python
df[["Tamb_K","T2_K","CompressorTempRise"]].head()
```

```python
plt.figure(figsize=(8,6))

plt.scatter(
    df["CompressorTempRise"],
    df["CompressorHealth"]
)

plt.xlabel("Compressor Temperature Rise")

plt.ylabel("Compressor Health")

plt.title("Compressor Temperature Rise vs Compressor Health")

plt.grid(True)

plt.show()
```

```python
df[["CompressorTempRise", "CompressorHealth"]].corr()
```

```python
# Combustor Pressure Loss
```

```python
df["CombustorPressureLoss"] = (df["P2_Pa"] - df["P3_Pa"]) / df["P2_Pa"]
```

```python
df[["P2_Pa", "P3_Pa", "CombustorPressureLoss"]].head()
```

```python
plt.figure(figsize=(8,6))

plt.scatter(
    df["CombustorPressureLoss"],
    df["CombustorHealth"]
)

plt.xlabel("Combustor Pressure Loss")
plt.ylabel("Combustor Health")
plt.title("Combustor Pressure Loss vs Combustor Health")

plt.grid(True)
plt.show()
```

# Phase 5 — Feature Validation

```python
new_features = [
    "CompressorPressureRatio",
    "CompressorTempRise",
    "CombustorPressureLoss",
    "CompressorHealth",
    "CombustorHealth"
]

df[new_features].corr()
```

```python
plt.figure(figsize=(8,6))

sns.heatmap(
    df[new_features].corr(),
    annot=True,
    cmap="coolwarm",
    fmt=".2f"
)

plt.title("Physics-Based Feature Correlation")

plt.show()
```

```python
target_cols = [
    "CompressorHealth",
    "CombustorHealth",
    "TurbineHealth",
    "OverallHealth",
    "Thrust_N",
    "TSFC_g_N_s"
]

sensor_cols = [
    "Altitude_m",
    "Mach",
    "Tamb_K",
    "Pamb_Pa",
    "RPM_rev_min",
    "FuelFlow_kg_s",
    "P2_Pa",
    "T2_K",
    "P3_Pa",
    "T3_K",
    "P4_Pa",
    "T4_K"
]

corr = df[sensor_cols + target_cols].corr()

plt.figure(figsize=(14,10))

sns.heatmap(
    corr.loc[sensor_cols, target_cols],
    annot=True,
    cmap="coolwarm",
    fmt=".2f"
)

plt.title("Sensor vs Target Correlation")

plt.show()
```
